```python
import numpy as np

arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)
print(type(arr1))
```

```python
arr2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(arr2)
print(type(arr2))
```

```python
arr3 = np.array([
    [[1, 2],
     [3, 4]],
    [[5, 6],
     [7, 8]]
])

print(arr3)
print(type(arr3))
```

```python
zeros = np.zeros(5)

print(zeros)
```

```python
ones = np.ones((2, 3))
print(ones)
```

```python
range_arr = np.arange(1, 10, 2)

print(range_arr)
```

```python
# 데이터 타입

정수 = np.array([1, 2, 3, 4, 5])
정수.dtype
```

```python
실수 = np.array([1.5, 2.5, 3.5])
실수.dtype
```

```python
# 배열 속성

arr1 = np.array([1, 2, 3, 4])

arr1.shape
```

```python
arr2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

arr2.shape
```

```python
arr3 = np.array([
    [[1, 2],
     [3, 4]],
    [[5, 6],
     [7, 8]]
])

arr3.shape
```

```python
# ndim

print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)
```

```python
# size - 배열의 총 요소 수를 알려준다

print(arr1.size)
print(arr2.size)
print(arr3.size)
```
