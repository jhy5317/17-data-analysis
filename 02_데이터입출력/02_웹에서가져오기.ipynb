```python
# 웹 스크래핑

from bs4 import BeautifulSoup
import requests
import re

agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/148.0.0.0 Whale/4.38.386.14 Safari/537.36"
headers = {"User-Agent": agent}

url = "https://en.wikipedia.org/wiki/List_of_American_exchange-traded_funds"

resp = requests.get(url, headers=headers)
soup = BeautifulSoup(resp.text, 'lxml')
rows = soup.select("div > ul > li")

# print("resp:", resp)
# print("=" * 100)
# print("resp.text:", resp.text)
# print("=" * 100)
# print("soup:", soup)
# print("=" * 100)
print("rows:", rows[50:55])
```

```python
etfs = {}

for row in rows:
    try:
        etf_name = re.findall(r"(.+?)\s\(", row.text)
        etf_market = re.findall(r"\((.+?):", row.text)
        etf_ticker = re.findall(r":\s(.+)\)", row.text)

        if (len(etf_ticker) > 0) & (len(etf_market) > 0) & (len(etf_name) > 0):
            etfs[etf_ticker[0]] = [etf_market[0], etf_name[0]]
    
    except AttributeError as err:
        pass

print(etfs)
```
