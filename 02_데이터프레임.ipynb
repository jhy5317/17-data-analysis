```python
import pandas as pd
```

```python
dict_data = {
    "c0": [1, 2, 3],
    "c1": [4, 5, 6],
    "c2": [7, 8, 9],
    "c3": [10, 11, 12],
    "c4": [13, 14, 15]
}

df = pd.DataFrame(dict_data)

print(type(df))
print(df)
```

```python
# pandas.DataFrame(2차원 배열,
#              index = 행 인덱스 배열,
#              columns =  열 이름 배열)

arr = [
    [15, "남", "덕영중"],
    [17, "여", "수리중"]
]

df = pd.DataFrame(arr, index = ["준서", "예은"], columns = ["나이", "성별", "학교"])

print(df)
```
