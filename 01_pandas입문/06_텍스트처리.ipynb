```python
import pandas as pd
import numpy as np
```

```python
ser = pd.Series(["Apple_사과", "Banana_바나나", "Cherry_체리", np.nan],
                index = ["First ", " Second", " Third", "Fourth"])
```

```python
result1 = ser.str.split("_")
result1
```

```python
result2 = ser.str.split("_", expand=True)
result2
```

```python
ser.str.split("_").str.get(1)
```

```python
# 문자열 메서드

idx = ser.index

print(idx)
print(idx.str.strip())
```

```python
print(ser.str.replace("_", ":", regex=False))
```

```python
print(ser.str.replace(r"[^a-zA-Z\s]", "", regex=True))
```

```python
print(ser.str[0])
```

```python
print(ser.str[0:4])
```

```python
ser.str.contains("A", na=False)
```

```python
ser.str.contains(r"[A|B][a-z]+")
```

```python
ser.str.fullmatch(r"[A|B][a-z]+")
```

```python
ser3 = pd.Series(["a1", "b2", "c3", "d4a5", "e6e7", "f8"], dtype="string")
ser3
```

```python
ser3.str.extractall(r"(\d)").groupby(level=0).agg("".join)
```

```python
ser3.str.extractall(r"([a-z])").groupby(level=0).agg("".join)
```

```python
# 숫자랑 필드랑 컬럼으로 분리

ser3.str.extract(r"(?P<Letter>[a-z]+)(?P<Digit>\d+)")
```
