```python
import pandas as pd
```

```python
dict_data = {
    "c0": [1, 2, 3],
    "c1": [4, 5, 6],
    "c2": [7, 8, 9],
    "c3": [10, 11, 12],
    "c4": [13, 14, 15]
}

df = pd.DataFrame(dict_data)

print(type(df))
print(df)
```

```python
# pandas.DataFrame(2차원 배열,
#              index = 행 인덱스 배열,
#              columns =  열 이름 배열)

arr = [
    [15, "남", "덕영중"],
    [17, "여", "수리중"]
]

df = pd.DataFrame(arr, index = ["준서", "예은"], columns = ["나이", "성별", "학교"])

print(df)
```

```python
print(df.index)
print(df.columns)
```

```python
# 행 인덱스 변경
# DataFrame객체.index = 행 인덱스 배열
# DataFrame객체.columns = 열 인덱스 배열

print(df)
```

```python
df.index = ["학생1", "학생2"]
print(df)
```

```python
df.columns = ["연령", "남녀", "소속"]
print(df)
```

```python
# 행 인덱스 변경 : DataFrame객체.rename(index = {기존 인덱스: 새 인덱스})
# 열 이름 변경 : DataFrame객체.rename(columns = {기존 이름: 새 이름})

matrix = [
    [15, "남", "덕영중"],
    [17, "여", "수리중"]
]

df = pd.DataFrame(matrix, index = ["준서", "예은"], columns = ["나이", "성별", "학교"])
print(df)
```

```python
df = df.rename(columns = {
    "나이": "연령",
    "성별": "남녀",
    "학교": "소속"
})
```

```python
df = df.rename(index = {
    "준서": "학생1",
    "예은": "학생2"
})

print(df)
```

```python
# 행/열 삭제

# 행 삭제 : DataFrame객체.drop(행 인덱스 또는 배열, axis = 0, axis = "index")
# 열 삭제 : DataFrame객체.drop(열 이름 또는 배열, axis = 1, axis = "columns")

# DataFrame() 함수로 데이터프레임 변환, 변수 df에 저장
exam_data = {
    "수학": [90, 80, 70],
    "영어": [98, 89, 95],
    "음악": [85, 95, 100],
    "체육": [100, 90, 90]
}

df = pd.DataFrame(exam_data, index = ["서준", "우현", "인아"])
df

df2 = df.copy()
df2 = df2.drop("우현")
df2
```

```python
df3 = df.copy()
df3 = df3.drop(["우현", "인아"], axis=0)
df3
```

```python
# 행, 열의 위치 바꾸기

# 행, 열 바꾸기: DataFrame객체.transpose() 또는 DataFrame.T
exam_data = {
    "이름": ["서준", "우현", "인아"],
    "수학": [90, 80, 70],
    "영어": [98, 89, 95],
    "음악": [85, 95, 100],
    "체육": [100, 90, 90]
}

df = pd.DataFrame(exam_data)

df = df.transpose()
df
```
