```python
# NaN값이 있는 시리즈 연산

s1 = pd.Series({"국어": np.nan, "영어": 80, "수학": 90})
s2 = pd.Series({"수학": 80, "국어": 90})

print(s1)
print(s2)

add = s1 + s2
sub = s1 - s2
mul = s1 * s2
div = s1 / s2
print(div)
print()

result = pd.DataFrame([add, sub, mul, div], index = ["덧셈", "뺄셈", "곱셈", "나눗셈"])
print(result)
```

```python
# 연산 메서드

s1 = pd.Series({"국어": np.nan, "영어": 80, "수학": 90})
s2 = pd.Series({"수학": 80, "국어": 90})

sr_add = s1.add(s2, fill_value=0)
sr_sub = s1.sub(s2, fill_value=0)
sr_mul = s1.mul(s2, fill_value=0)
sr_div = s1.div(s2, fill_value=0)

result = pd.DataFrame([sr_add, sr_sub, sr_mul, sr_div],
                      index = ["덧셈", "뺄셈", "곱셈", "나눗셈"])

print(result)
```
